Using an automatic differentiation system like PyTorch Autograd:

1. Create a tensor x with requires_grad = True

2. Define y1 = x ** 2

3. Define y2 = 4 * x

4. Define L = y1 + y2

5. Call L.backward()

6. Read the gradient from x.grad

What Autograd does internally:

* Builds the same computation graph you drew in part (a)

* Applies the chain rule automatically during backward()

* Accumulates gradients flowing back to x

The final value stored in x.grad will be 10, matching the manual result.

# Question 04: [ Marks 15 ]

## Write Answer 04:

```python
import torch

x = torch.tensor(3.0, requires_grad=True)
y = x**2 + 4 * x

y.backward()
print(x.grad)
```

# Question 05: [ Marks 10 ]

Create two tensors:
w = 2.0 (requires_grad=True) <br>
b = 1.0 (requires_grad=True) <br>

Let input x = 3.0. <br>
Compute z = w*x + b. <br>
Use Sigmoid to get y_pred.
 <br>

Define Binary Cross Entropy manually (without built-in loss). <br>
Compute gradients using backward(). <br>
Print gradients of w and b. <br>

Write full code.

## Write Answer 05:

```python
import torch

w = torch.tensor(2.0, requires_grad=True)
b = torch.tensor(1.0, requires_grad=True)

x = torch.tensor(3.0)
y_true = torch.tensor(1.0)  # binary label

z = w * x + b

y_pred = torch.sigmoid(z)

loss = -(y_true * torch.log(y_pred) + (1 - y_true) * torch.log(1 - y_pred))

loss.backward()

print("Gradient of w:", w.grad)
print("Gradient of b:", b.grad)
```

# Question 07: [ Marks 20 ]


Create tensor w = 2.0 (requires_grad=True). <br>

Print w.grad. <br>

Do not clear gradients. <br>
Explain what you observe. <br>

## Write Answer 07:



```python
import torch

w = torch.tensor(2.0, requires_grad=True)

loss1 = w ** 2
loss1.backward()

loss2 = 3 * w
loss2.backward()

print(w.grad)
```

**Explanation:**

After the first backward(), w.grad = 4

After the second backward(), PyTorch adds the new gradient to the existing one

So final gradient is 4 + 3
