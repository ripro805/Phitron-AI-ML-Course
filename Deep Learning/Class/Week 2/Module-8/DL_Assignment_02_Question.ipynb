# Question 02: [ Marks 10 ]

What is forward propagation in an artificial neural network? Explain step by step how data moves from the input layer to the output layer.

# Forward Propagation in Artificial Neural Networks

Forward propagation is the process by which **input data is passed through the neural network layer by layer to produce an output**.  
In this process, **no learning happens**. The network only performs computations using the current weights and biases.



## Step-by-Step Flow of Forward Propagation

Consider a neural network with:  
- Input layer  
- One or more hidden layers  
- Output layer  



### Step 1: Input Layer

Let the input feature vector be:

$x = (x_1, x_2, ..., x_n)$

The input layer **does not perform any computation**. It only passes the input values to the next layer.



### Step 2: Weighted Sum in a Hidden Neuron

For a neuron $j$ in a hidden layer:

- Weights: $w_j = (w_{j1}, w_{j2}, ..., w_{jn})$  
- Bias: $b_j$

The linear combination is computed as:

$z_j = \sum_{i=1}^{n} w_{ji} x_i + b_j$



### Step 3: Activation Function

The weighted sum $z_j$ is passed through an **activation function** $g$:

$a_j = g(z_j)$

Common activation functions:  
- Sigmoid  
- ReLU  
- Tanh  

This step introduces **non-linearity** into the network.



### Step 4: Forward Propagation Through Hidden Layers

The output of one layer becomes the input to the next layer.

For layer $l$:

$z^{(l)} = W^{(l)} a^{(l-1)} + b^{(l)}$

$a^{(l)} = g(z^{(l)})$

This process continues until the final layer.



### Step 5: Output Layer Computation

At the output layer:

$z^{(L)} = W^{(L)} a^{(L-1)} + b^{(L)}$

The output activation depends on the task:

- Regression: $\hat{y} = z^{(L)}$  
- Binary Classification: $\hat{y} = \sigma(z^{(L)})$  
- Multi-class Classification: $\hat{y} = \text{softmax}(z^{(L)})$  



### Step 6: Final Output

The final output $\hat{y}$ or $\hat{\mathbf{y}}$ is the **prediction of the neural network**.  

This output is later used to **compute the loss** during training.

# Why Do We Need Activation Functions in Neural Networks?

Activation functions are crucial in neural networks because they introduce **non-linearity** into the model. Without them, neural networks cannot learn complex patterns.



## 1. Purpose of Activation Functions

- **Non-linearity:** Real-world data often has complex, non-linear relationships. Activation functions allow the network to approximate these relationships.  
- **Decision boundaries:** They enable the network to create non-linear decision boundaries.  
- **Enabling deep learning:** Without activation functions, multiple layers would collapse into a single linear transformation, making deep networks no more powerful than a single layer.



## 2. What Happens If We Only Use Linear Transformations?

Suppose each layer performs only a linear transformation:

$$
a^{(l)} = W^{(l)} a^{(l-1)} + b^{(l)}
$$

For a network with three layers:

$$
\begin{aligned}
a^{(1)} &= W^{(1)} x + b^{(1)} \\
a^{(2)} &= W^{(2)} a^{(1)} + b^{(2)} \\
a^{(3)} &= W^{(3)} a^{(2)} + b^{(3)}
\end{aligned}
$$

We can combine all layers into a single linear function:

$$
a^{(3)} = (W^{(3)} W^{(2)} W^{(1)}) x + (W^{(3)} W^{(2)} b^{(1)} + W^{(3)} b^{(2)} + b^{(3)})
$$

**Consequences:**

- The network can only model **linear relationships**.  
- Adding more layers **does not increase expressive power**.  
- The network cannot capture non-linear patterns in data.



## 3. Common Activation Functions

- **Sigmoid:** $\sigma(x) = \frac{1}{1 + e^{-x}}$  
- **ReLU:** $\text{ReLU}(x) = \max(0, x)$  
- **Tanh:** $\tanh(x)$

# Sigmoid Activation Function

The **sigmoid activation function** is one of the most commonly used activation functions in neural networks. It maps any real-valued input to a value between 0 and 1.



## 1. Mathematical Formula

The sigmoid function is defined as:

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

Where:  
- $x$ is the input to the neuron  
- $\sigma(x)$ is the output of the sigmoid function



## 2. Properties of Sigmoid

- Output range: $0 < \sigma(x) < 1$  
- Smooth and differentiable everywhere  
- Non-linear, so it allows the network to learn complex patterns  
- Derivative:

$$
\sigma'(x) = \sigma(x) \cdot (1 - \sigma(x))
$$

This derivative is useful during backpropagation.



## 3. Where Is It Used?

- **Binary classification:** As an output activation for predicting probabilities  
- **Logistic regression:** Naturally produces outputs in [0, 1]  
- **Early neural networks:** Often used in hidden layers  



## 4. Advantages

- Smooth gradient helps with learning  
- Output can be interpreted as **probability**  
- Non-linear, so it can model complex patterns  



## 5. Disadvantages / Problems

1. **Vanishing Gradient Problem:**  
   - For very large positive or negative inputs, $\sigma(x)$ saturates near 1 or 0  
   - Its derivative becomes very small:  

   $$
   \sigma'(x) = \sigma(x)(1 - \sigma(x)) \approx 0
   $$

   - This slows down or stops learning in deep networks

2. **Not zero-centered:**  
   - Output is always positive (0 to 1)  
   - Can cause **zig-zagging** of gradients during optimization

3. **Computationally expensive** compared to ReLU

# Softmax Activation Function

The **softmax activation function** is used to convert raw scores (logits) from a neural network into **probabilities** for multiple classes.  

It is commonly used in **multi-class classification problems**.



## 1. Mathematical Formula

For a vector of inputs $z = [z_1, z_2, ..., z_K]$, the softmax output for class $i$ is:

$$
\text{softmax}(z_i) = \frac{e^{z_i}}{\sum_{j=1}^{K} e^{z_j}}
$$

Where:  
- $K$ = total number of classes  
- $z_i$ = input score (logit) for class $i$  
- The output is a probability: $0 < \text{softmax}(z_i) < 1$  
- All outputs sum to 1:

$$
\sum_{i=1}^{K} \text{softmax}(z_i) = 1
$$



## 2. Why Softmax Is Used in Multi-Class Classification

- It converts **logits into probabilities** for each class  
- The class with the **highest probability** is chosen as the predicted class  
- Ensures that outputs are **mutually exclusive**, i.e., only one class can have the highest probability



## 3. Difference Between Softmax and Sigmoid

| Feature              | Sigmoid                             | Softmax                               |
|----------------------|------------------------------------|--------------------------------------|
| Output range          | 0 to 1                             | 0 to 1                               |
| Number of outputs     | 1 (binary probability)             | K (probabilities for K classes)      |
| Sum of outputs        | Not necessarily 1                  | Sum = 1                              |
| Use case              | Binary classification               | Multi-class classification            |
| Activation type       | Scalar per neuron                  | Vector over all classes               |

# Structure of an Artificial Neural Network (ANN)

An **Artificial Neural Network (ANN)** is a computational model inspired by the human brain. It is designed to recognize patterns and learn complex relationships from data.

A typical ANN consists of:

- **Input layer**  
- **One or more hidden layers**  
- **Output layer**

These layers are connected via **weights** and **biases**, and each neuron uses an **activation function**.



## 1. Input Layer

- The input layer receives the **raw input features** from the dataset.  
- It **does not perform any computation**. Its main job is to pass inputs to the next layer.  

Example:

If we have 3 features: `x1, x2, x3`, the input vector is:

$$
\mathbf{x} = [x_1, x_2, x_3]
$$



## 2. Hidden Layers

- Hidden layers perform computations on the inputs received from the previous layer.  
- Each neuron computes a **weighted sum** of its inputs plus a bias, then applies an **activation function**:

$$
z_j^{(l)} = \sum_{i=1}^{n} w_{ji}^{(l)} a_i^{(l-1)} + b_j^{(l)}
$$

$$
a_j^{(l)} = g(z_j^{(l)})
$$

Where:  
- $w_{ji}^{(l)}$ = weight from neuron $i$ in layer $(l-1)$ to neuron $j$ in layer $l$  
- $b_j^{(l)}$ = bias of neuron $j$ in layer $l$  
- $g$ = activation function (e.g., ReLU, Sigmoid, Tanh)  
- $a_i^{(l-1)}$ = output from previous layer

**Purpose of hidden layers:**  

- Introduce **non-linearity** to the network  
- Enable the network to learn **complex patterns** in the data  



## 3. Output Layer

- The output layer produces the **final prediction** of the network.  
- The activation function depends on the task:  

  - **Regression:** Linear activation  
  - **Binary classification:** Sigmoid activation  
  - **Multi-class classification:** Softmax activation

$$
\hat{y} = g_{\text{output}}(W^{(L)} a^{(L-1)} + b^{(L)})
$$

Where $L$ is the index of the output layer.



## 4. How Weights, Biases, and Activation Functions Work Together

1. **Weights ($W$):**  
   - Control the strength of connections between neurons  
   - Adjusted during training to minimize the loss function

2. **Biases ($b$):**  
   - Allow the activation function to **shift**  
   - Help the network fit data better

3. **Activation Functions ($g$):**  
   - Introduce **non-linearity**  
   - Enable the network to learn complex, non-linear relationships

Together, they transform the input vector $x$ into the predicted output $\hat{y}$:

$$
x \xrightarrow{\text{weights, biases, activation}} a^{(1)} \xrightarrow{\text{weights, biases, activation}} \dots \xrightarrow{\text{output layer}} \hat{y}
$$



## 5. Simple Example

Suppose we have a small network:

- **Input layer:** 2 features ($x_1, x_2$)  
- **Hidden layer:** 2 neurons with ReLU activation  
- **Output layer:** 1 neuron with Sigmoid activation

Let:

$$
W^{(1)} =
\begin{bmatrix} 0.5 & -0.3 \\ 0.8 & 0.2 \end{bmatrix}, \quad
b^{(1)} = [0.1, -0.1]
$$

$$
W^{(2)} = [0.7, -0.5], \quad b^{(2)} = 0.2
$$

**Step 1: Hidden layer calculations**

$$
z_1^{(1)} = 0.5 x_1 - 0.3 x_2 + 0.1, \quad a_1^{(1)} = \text{ReLU}(z_1^{(1)})
$$

$$
z_2^{(1)} = 0.8 x_1 + 0.2 x_2 - 0.1, \quad a_2^{(1)} = \text{ReLU}(z_2^{(1)})
$$

**Step 2: Output layer calculation**

$$
z^{(2)} = 0.7 a_1^{(1)} - 0.5 a_2^{(1)} + 0.2
$$

$$
\hat{y} = \sigma(z^{(2)})
$$

The final $\hat{y}$ is the network’s **prediction**.
