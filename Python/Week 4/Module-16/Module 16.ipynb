#RegPlot and lmplot

regression plot-->Regression plot হলো এমন একটা চার্ট যা দুইটা numeric variable-এর মধ্যে সম্পর্ক (relationship) দেখায়,

#Pairplot

sns.pairplot() হলো Seaborn-এর একটা Figure-level function,
যা একাধিক numeric variable এর মধ্যে pairwise relationships দেখায় —
অর্থাৎ, প্রতিটি ভ্যারিয়েবলের সাথে প্রতিটি ভ্যারিয়েবলের সম্পর্ক আলাদা scatter plot আকারে দেখায়।

```python
student
```

```python
student_marks = student[['Marks','study_hours','attendance_rate','gender']]
student_marks
```

```python
sns.pairplot(data=student_marks)
```

```python
sns.pairplot(data=student_marks, kind='hist')
```

```python
sns.pairplot(data=student_marks,hue='gender')
```

#Joint Plot

sns.jointplot() হলো Seaborn-এর Figure-level function,
যা দুইটা numeric variable এর মধ্যে relationship (scatter) এবং
distribution (histogram/KDE) — দুটোই একসাথে দেখায়।

```python
sns.jointplot(data=student_marks , x ='study_hours' , y='Marks')
```

```python
sns.jointplot(data=student_marks , x ='study_hours' , y='Marks',hue='gender')
```

```python
sns.jointplot(data=student_marks , x ='study_hours' , y='Marks',kind='kde')
```
