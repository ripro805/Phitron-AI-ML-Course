```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace", "Hannah","Sakib"],
    "City": ["New York", "Los Angeles", "Newark", "Boston", "New Delhi", "Chicago", "New Orleans", "Houston","H Los Ang"],
    "Department": ["HR", "IT", "Finance", "IT", "HR", "Marketing", "Finance", "HR", "HR"],
    "Salary": [50000, 60000, 55000, 70000, 52000, 58000, 62000, 51000,70000]
}

df = pd.DataFrame(data)
df
```

```python
df.loc[df['City'].str.contains('New')]
```

```python
#without case sensivity
df.loc[df['City'].str.contains('New',case=False)]
```

```python
#starts with los
df.loc[df['City'].str.contains(r"^Los")]
```

```python
#ends with rk
df.loc[df['City'].str.contains(r"rk$")]
```

```python
#starts with vowel
df.loc[df['Name'].str.contains(r'^[aeiouAEIOU]')]
```

```python
# string contains New or Los

df.loc[df['City'].str.contains(r"New|Los")]
```

```python
#min max scaling

mn = df['Total Marks'].min()
mx = df['Total Marks'].max()


df1['Scaled Marks'] = df['Total Marks'].apply(lambda x : (x-mn)/(mx-mn))
df1
```

```python
# custom built function

def grading_system(marks):

    if marks >=260:
        return 'A+'
    elif marks>=250:
        return 'A'
    else:
        return 'A-'


df1['Grade'] = df1['Total Marks'].apply(grading_system)
df1
```

```python
def marking_system(df1):
    a = df1['Data Structure Marks'] * 2
    b = df1['Python Marks'] * 3
    c = df1['Algorithm Marks'] * 4

    return a+b+c



df1['Exceptional Marks'] = df1.apply(marking_system,axis=1)
df1
```

```python
#https://leetcode.com/problems/delete-duplicate-emails/?envType=study-plan-v2&envId=30-days-of-pandas&lang=pythondata
import pandas as pd

def delete_duplicate_emails(person: pd.DataFrame) -> None:
    person.sort_values('id',inplace=True)
    person.drop_duplicates(subset=['email'],inplace=True)
```

```python
#https://leetcode.com/problems/count-salary-categories/description/?envType=study-plan-v2&envId=30-days-of-pandas&lang=pythondata
import pandas as pd

def count_salary_categories(accounts: pd.DataFrame) -> pd.DataFrame:
    low_salary=(accounts['income']<20000).sum()
    average_salary=((accounts['income']>=20000) & (accounts['income']<50000)).sum()
    high_salary=(accounts['income']>50000).sum()

    data=[('Low Salary',low_salary),('Average Salary',average_salary),('High Salary',high_salary)]

    df=pd.DataFrame(data,columns=['category','accounts_count'])
    return df
```
