```python
data = {
    'Product': ['Pen', 'Book', 'Eraser', 'Pencil'],
    'Price': [10, 50, 5, 7],
    'Quantity': [100, 60, 200, 150]
}

df=pd.DataFrame(data)
print(df.columns)
print(df.index)
print("\nA single column : \n",df['Product'])
print("\nA full Row : \n",df.iloc[0])
```

```python
#4
f_r_loc=df.loc[0]
f_r_iloc=df.iloc[0]
print("First row (.loc):\n", f_r_loc)
print("\nFirst row (.iloc):\n", f_r_iloc)

s_c_loc=df.loc[:,'Price']
s_c_iloc=df.iloc[:,1]
print("\nSecond column (.loc):\n", s_c_loc)
print("\nSecond column (.iloc):\n", s_c_iloc)

block_loc = df.loc[1:2, ['Price', 'Quantity']]
block_iloc = df.iloc[1:3, 1:3]
print("\n2x2 block (.loc):\n", block_loc)
print("\n2x2 block (.iloc):\n", block_iloc)
```

```python
data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Math": [85, 90, 78, 92],
    "English": [88, 75, 82, 95]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

df_drop_col = df.drop(columns=['English'])
print("\nAfter dropping 'English' column:\n", df_drop_col)

df_drop_row = df.drop(index=2)
print("\nAfter dropping row with index 2:\n", df_drop_row)
```

```python

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Department": ["HR", "IT", "Finance", "IT", "HR"],
    "Salary": [50000, 60000, 55000, 60000, 52000]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# Sort by Salary (ascending) and Name (descending)
sorted_df = df.sort_values(by=["Salary", "Name"], ascending=[True, False])
print("\nSorted DataFrame:\n", sorted_df)
```

```python
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace", "Helen"],
    "Score": [85, 55, 30, 70, 90, 40, 25, 65]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# Students with Score > 60
high_scores = df[df["Score"] > 60]
print("\nStudents with Score > 60:\n", high_scores)

# Students with Score < 40
low_scores = df[df["Score"] < 40]
print("\nStudents with Score < 40:\n", low_scores)
```
