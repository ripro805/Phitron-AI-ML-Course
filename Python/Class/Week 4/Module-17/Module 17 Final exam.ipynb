```python
trning_hrs=df['TrainingHours'].astype(float).to_numpy()
mean=np.mean(trning_hrs)
std_dev=np.std(trning_hrs)
print(mean)
print(std_dev)
```

#Question 2

```python
for dept,grp in df.groupby('Department'):
  median_salary=grp['Salary'].median()
  df.loc[(df["Department"]==dept)& (df['Salary'].isna()),'Salary']=median_salary

  fltrd=df[(df['ExperienceYears']>12) & (df['ProjectCount']>5)]
tot_sal=fltrd['Salary'].sum()
print(tot_sal)
```

#question 3


```python
filtrd=df[
    ((df['Department']=='IT')| (df['Department']=='Finance'))&
    ((df['Age']>=30) & df['Age']<=45) &
    (df["PerformanceScore"] > 88)
          ]
srtd=filtrd.sort_values(by='Salary',ascending=False)
res = srtd[["Name", "Department", "Salary", "PerformanceScore"]]
print(res)
```
